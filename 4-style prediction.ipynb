```python
# 图片预处理
tf = trn.Compose([
    trn.Resize([224, 224]),
    trn.ToTensor(),
    trn.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])])
```

# 训练

## 设置

```python
model = models.densenet121(pretrained=True)
# 修改模型，与任务相关
model.classifier = nn.Linear(in_features=1024, out_features=13, bias=True)
# cuda 并行
model = nn.DataParallel(model)
```

```python
# 定义损失函数
criterion = nn.CrossEntropyLoss().cuda()
# 定义优化器
optimizer = torch.optim.SGD(model.parameters(),
                            0.001, # learning rate 最重要的超参数之一
                            momentum = 0.9,
                            weight_decay = 1e-4)
```

```python
# 加载模型
######和之前a1中训练不一样的第一处
checkpoint = torch.load("/content/gdrive/MyDrive/T_ML/0-try1-13styles/best_checkpoint5311_50_49.pth.tar")
model.load_state_dict(checkpoint['state_dict'])
optimizer.load_state_dict(checkpoint['optimizer'])

# 设置模型为评估模式
model.eval()
```

```python
import pandas as pd
```

## 进行预测

```python
# 遍历b文件夹中的图片，并进行预测
image_folder = '/content/gdrive/MyDrive/T_ML/img-mht/mht'  # 需要预测的新文件夹的路径
predictions = []

for i, image_file in enumerate(os.listdir(image_folder)):
    if i == 45437:
      break
    image_path = os.path.join(image_folder, image_file)

    # 加载图像并进行预处理
    image = Image.open(image_path)
    image = tf(image).unsqueeze(0)

    # 使用模型进行预测
    with torch.no_grad():
        outputs = model(image)
        _, predicted_idx = torch.max(outputs, 1)
        predicted_label = predicted_idx.item()

    # 将预测结果添加到列表中
    predictions.append({'image': image_file, 'label': predicted_label})
# 创建DataFrame保存预测结果
df_predictions = pd.DataFrame(predictions)

# 打印DataFrame
print(df_predictions)
df_predictions.to_csv('df_predictions45437-1.csv', index=False)
```

```python
# 读取csv文件
df = pd.read_csv('df_predictions45437-1.csv')

# 查找包含(1)的行并删除
df = df[~df['image'].str.contains('\(1\)')]

# 保存修改后的文件
df.to_csv('df_predictions45437-2.csv', index=False)
```
