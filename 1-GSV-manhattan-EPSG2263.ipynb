```python
#对于一条路来说的

def splitToSegment(road):
    segments=[]
    for i in range(len(road.xy[0])-1):   #road.xy[0]是第一个array的长度
        segments.append(LineString([(road.xy[0][i],road.xy[1][i]),(road.xy[0][i+1],road.xy[1][i+1])]))
    return segments
```

```python
#循环把所有的路都转换为线段

segmentList = []

for g  in roads.geometry:
    segmentList.extend(splitToSegment(g))
```

## 3 建筑物立面中点投影配对数据——立面中点投影到街道上，获取投影点

```python
bvld_mht1 = bvld_mht.loc[:,['doitt_id','geometry','cnstrct_yr']]
bvld_mht1
```

```python
display(bvld_mht1.head(2))
display(segmentDF.head(2))
```

```python
#转换为以米为单位的好计算

bvld_mht1 = bvld_mht1.to_crs("epsg:2263")
segmentDF['road'] = gpd.GeoSeries(segmentDF['road']).set_crs("epsg:4326").to_crs("epsg:2263")
```

```python
display(bvld_mht1.head(2))
display(segmentDF.head(2))
```

```python
#将建筑的坐标点单独保存
bvld_p = bvld_mht1['geometry'].reset_index(drop=True)
bvld_p
```

```python
#换个和样例代码里一样的名字

midFacades = bvld_p
print(midFacades[17])
```

```python
#获取建筑物点-投影点 组合

midFacades = bvld_p
streetPts = []

roads = segmentDF.road.tolist()

# 创建rtree
street_tree = index.Index()
for idx, g in enumerate(roads):
    street_tree.insert(idx, g.bounds)
    
# 对于所有中点，获取最近的10个道路
for mid in midFacades:
    minDist = 10000
    streetPt = None
    street_idxs = list(street_tree.nearest((mid.xy[0][0], mid.xy[1][0], mid.xy[0][0], mid.xy[1][0]), 10))

    # 对于每条道路，获取离建筑物中点最近的一个点
    for street_idx in street_idxs:
        # 获取道路
        line = roads[street_idx]
        # 获取最近点
        nearPt = nearest_points(mid, line)[1]
        # 计算距离
        dist = nearPt.distance(mid)
        # 判断距离大小
        if dist < minDist:
            minDist = dist
            streetPt = nearPt

    streetPts.append(streetPt)
```

```python
len(midFacades), len(streetPts)
```

```python
print(midFacades[17])
print(streetPts[17])
```

## 4 计算角度

```python
angles = []

for pt1, pt2 in zip(midFacades, streetPts):
    angle_de = np.degrees(np.arctan2(pt1.y - pt2.y, pt1.x - pt2.x))
    # 坐标系的转换
    angles.append(((((angle_de + 360) % 360) * -1 + 90 + 360) % 360).astype(int))
```

```python
#换回4326

# 将midFacades转换为epsg:4326
midFacades = gpd.GeoSeries(midFacades).set_crs("epsg:2263", allow_override=True).to_crs('epsg:4326')

# 将streetPts转换为epsg:4326
streetPts = gpd.GeoSeries(streetPts).set_crs("epsg:2263", allow_override=True).to_crs('epsg:4326')
```

```python
print(midFacades[17])
```

```python
#把上面的结果加入bvld_mht1
bvld = bvld_mht1
bvld.head(2)
```

```python
bvld = bvld.to_crs("epsg:4326")

bvld['mid_x'] = [pt.xy[0][0] for pt in midFacades]
bvld['mid_y'] = [pt.xy[1][0] for pt in midFacades]
bvld['streetPt_x'] = [pt.xy[0][0] for pt in streetPts]
bvld['streetPt_y'] = [pt.xy[1][0] for pt in streetPts]

bvld['angle'] = angles
```
